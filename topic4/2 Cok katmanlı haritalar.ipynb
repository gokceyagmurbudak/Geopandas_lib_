```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
malmo_mahalle= gpd.read_file("malmo-pop.shp")
malmo_yollar = gpd.read_file("malmo-streets.shp")
malmo_anaokulları = gpd.read_file("preschools.shp")
```

### Yan yana Grafikler

```python
# fig görüntüyü ax çerçeve
fig , ax= plt.subplots(3,1,figsize= (14,12))
```

```python
fig , ax= plt.subplots(1,3,figsize= (14,12))
```

```python
fig , ax= plt.subplots(1,3,figsize= (20,18)); # 1 satır 3 sütun
malmo_mahalle.plot(ax=ax[0],figsize= (18,14));
malmo_yollar.plot(ax=ax[1],figsize= (18,14));
malmo_anaokulları.plot(ax=ax[2],figsize= (18,14));
```

# Üst üste bindirilen çizimler (Çok katmanlı- Overlay)
#### Üst üste bindirme işlemi yapabilmek için 3 haritanın da aynı coğrafi referans coordinat düzleminde (CRS) olması gerekmektedir.

```python
fig, ax= plt.subplots(1,1, figsize=(12,10))
malmo_yollar.plot(ax=ax, color="blue",linewidth=0.3, alpha=0.5);
malmo_mahalle.plot(ax=ax, color="purple", alpha=0.7);
malmo_anaokulları.plot(ax=ax, color="yellow",figsize=(12,12),markersize=40,edgecolor="white", alpha=0.8,marker="o");

plt.title("İsveç Malmo Mahalleleri", fontsize=30, fontname="DejaVu Sans", color="grey")
ax.axis("off")
plt.axis("equal")
plt.show()
```

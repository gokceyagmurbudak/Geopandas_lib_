```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
malmo_mahalle = gpd.read_file("malmo-pop.shp")
malmo_mahalle.head()
```

```python
malmo_mahalle.shape  # (202, 21) 202 mahalle , 21 sütunu var
```

```python
malmo_mahalle.crs # Coğrefi verimiz(dataframe) epsg:3006,catersian (meter), method: Transverse Mercator, koordinat sisteminde
```

```python
malmo_mahalle.plot();
```

```python
malmo_mahalle.plot(figsize=(12,10));
```

```python
malmo_mahalle.plot(figsize=(12,10),
                  color = "purple");
```

```python
# alpha şeffaflık gösterir bize ve 0 ile 1 arasındadır .1 mattır ve 0 a gidildikçe şeffaflık kazanır.
malmo_mahalle.plot(figsize=(12,10),
                  color = "purple",
                  alpha= 0.7);
```

```python
# Mahalle sınırlarını belirlemek için 'edgecolor' parametresi kullanılır.
# edgecolor rengini belirlemek için HEX kodunu kullanabiliriz ve 
# bu adresten kodunu alabiliriz :
# https://www.google.com/search?q=color+picker&oq=color+picker&aqs=chrome..69i57.3986j0j7&sourceid=chrome&ie=UTF-
malmo_mahalle.plot(figsize=(12,10),
                  color = "purple",
                  alpha = 0.7,
                  edgecolor = '#3fe8d4' );
```

```python
# Mahalle sınır çizgilerinin kalınlığını 'linewidth' methoduyla ayarlayabiliriz.
malmo_mahalle.plot(figsize=(12,10),
                  color = "purple",
                  alpha = 0.7,
                  edgecolor = '#3fe8d4',
                  linewidth = 2);
```

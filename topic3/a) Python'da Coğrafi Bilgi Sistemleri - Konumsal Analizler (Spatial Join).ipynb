```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
shp = "countries.shp"
ulkeler= gpd.read_file(shp)
ulkeler.head()
```

```python
catısmalar = gpd.read_file("acled2019.shp")
catısmalar.head()
```

```python
ulkeler_geometry= ulkeler[["NAME","geometry"]]
ulkeler_geometry.head()
```

```python
ulkeler_isim= ulkeler[["NAME","POP_EST"]]
ulkeler_isim.head()
```

### Name sütununu ortak kullanarak tabloları bileştireceğiz.

```python
ulkeler_gdf_birlesik=ulkeler_geometry.merge(ulkeler_isim, on= "NAME" , how="left")
ulkeler_gdf_birlesik.head()
```

```python
type(ulkeler_gdf_birlesik)
```

```python
ulkeler_df_birlesik=ulkeler_isim.merge(ulkeler_geometry, on= "NAME", how = "left")
ulkeler_df_birlesik.head()
```

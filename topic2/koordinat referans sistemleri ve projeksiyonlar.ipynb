# Koordinat Referans Sistemleri (CRS) ve Projeksiyonlar

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
shp = "countries.shp"
ulkeler = gpd.read_file(shp)
ulkeler.head()
```

```python
ulkeler.crs
```

```python
catısmalar = gpd.read_file("acled2019.shp", rows=500)
catısmalar.head()
```

```python
catısmalar.crs
```

```python
ulkeler.geometry[:5] # ulkelerin geometry sütununu inceledik
```

```python
# epsg:3395 (mercator proj.) kullanacağız. bu da kuzey ve güneye gidildikçe alan genişlemesi yaşayacağımız anlamına geliyor.
# bu yüzden antartika kıtası olanları veri setimizde kullanmak istemiyoruz.

ulkeler = ulkeler[(ulkeler.NAME!= "Antarctica") & (ulkeler.NAME!= "Fr. S. Antarctic Lands")]  # "&" sembol "and" demektir

ulkeler_projected=ulkeler.to_crs(epsg=3395) # projectionu değiştirdik.

ulkeler_projected.geometry[:5] #geometry'de projection değişimiyle birlikte, koordinatlarımızın değişimini görüyoruz
```

```python
ulkeler.plot();
ulkeler_projected.plot();
```

```python
fig, ax = plt.subplots()
ulkeler_projected.plot(ax=ax)
catısmalar.plot(ax=ax,color="red") # tek noktada toplanmasının sebebi farklı projeksiyon sistemlerinde olmaları.
```

```python
fig, ax= plt.subplots()
ulkeler.plot(ax=ax)
catısmalar.plot(ax=ax,color="red") # ikisi de aynı projeksiyon sistemleri olduğu için doğru sonuç elde ettik.(epsg 4326)
```

```python
catısmalar_projected=catısmalar.to_crs(epsg= 3395)
fig, ax= plt.subplots()
ulkeler_projected.plot(ax=ax)
catısmalar_projected.plot(ax=ax, color="red")# ikisi aynı projeksiyon sistemleri olduğu için doğru sonuç elde ettik.(epsg 3395)
```
